# COVID-19 Lung CT Scan Image Classification Model
In this notebook, I will create a Keras Sequential Model to classify images of lung CT scan into 2 classes: 0 (negative to COVID-19) and 1 (positive to COVID-19). [ADD MORE]

```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
import keras
from keras_tuner.tuners import RandomSearch
import pandas as pd
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
import os
```

# Loading the Data

```python
data = tf.keras.utils.image_dataset_from_directory('data')
```

```python
iterator = data.as_numpy_iterator()
```

```python
batch = iterator.next()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
  ax[idx].imshow(img.astype(int))
  ax[idx].title.set_text(batch[1][idx])
```

Above is a batch of data where 0 represents the CT scan of a patient that is negative to COVID-19 and 1 represents the CT scan of a patient that is positive to COVID-19.

# Preprocessing 

```python
data = data.map(lambda x,y: (x/255, y))
```

We've normalized the data so that they are between 0 and 1.

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2) + 1
test_size = int(len(data)*.1) + 1
```

```python
print('The data contains a total of ' + str(train_size + val_size + test_size) + ' batches.')
print('The training set includes ' + str(train_size) + ' batches.')
print('THe validation set includes ' + str(val_size) + ' batches.')
print('The test set includes ' + str(test_size) + ' batches.')
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

The training set includes roughly 70% of the data. The validation set includes roughly 20% of the data. The test set includes roughly 10% of the data.

# Hyper-paremeter Tuning 

```python
def build_model(hp):
    model = keras.Sequential()

    num_conv_layers = hp.Int('num_conv_layers', 1, 3)
    for i in range(num_conv_layers):
        model.add(keras.layers.Conv2D(
            filters=hp.Int('filters_' + str(i), 32, 256, step=32),
            kernel_size=hp.Choice('kernel_size_' + str(i), values=[3, 5]),
            activation='relu',
            ))
        model.add(keras.layers.MaxPooling2D(pool_size=2))

    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(256, activation='relu'))

    model.add(keras.layers.Dense(1, activation='sigmoid'))

    optimizer = hp.Choice('optimizer', values=['adam', 'rmsprop', 'sgd'])

    lr = hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='log')

    if optimizer == 'adam':
        optimizer = keras.optimizers.Adam(learning_rate=lr)
    elif optimizer == 'rmsprop':
        optimizer = keras.optimizers.RMSprop(learning_rate=lr)
    else:
        optimizer = keras.optimizers.SGD(learning_rate=lr)

    model.compile(optimizer=optimizer,
                  loss=tf.losses.BinaryCrossentropy(),
                  metrics=['accuracy'])
    return model
```

Above the build_model function that tunes the number of layers, layer settings, optimizer, and learning rate to find the best hyper-parameters.

```python
tuner = RandomSearch(
    build_model, 
    objective='val_accuracy', 
    max_trials=5, 
    directory='tuner_results',
    project_name='covid_tuner')
```

```python
tuner.search(train, epochs=20, validation_data=(val))
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_model = tuner.hypermodel.build(best_hps)
```

After finding the best parameters, the model's settings are adjusted to include those parameters.

# Training the Model

```python
log_dir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
```

```python
hist = best_model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
best_model.summary()
```

Above is the summary of the model's layers, output shapes, and parameter numbers.

# Model Performance

```python
pd.DataFrame(hist.history)
```

Above is the data frame that shows the loss on the training set, accuracy on the training set, loss on the validation set, and accuracy on the validation set through out the 20 epochs.

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

Above is the graph showing the loss on the training set and the loss on the validation set through out the 20 epochs.

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

Above is the graph showing the accuracy on the training set and the loss on the validation set through out the 20 epochs.

# Evaluating the Model

```python
precision = Precision()
recall = Recall()
accuracy = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    pred = best_model.predict(X)
    precision.update_state(y, pred)
    recall.update_state(y, pred)
    accuracy.update_state(y, pred)
```

```python
print('Precision: ' + str(precision.result()))
print('Recall: ' + str(recall.result()))
print('Accuracy: ' + str(accuracy.result()))
```

The precision score of the model on the test set is 100%. The recall score of the model on the test set is 96.9%. The accuracy score of the model on the test set is 98.6%.
